#### Handling Missing Values

```python
house_pricing.isna().sum()
```

```python
#Dropping the column no of times visited
house_pricing.drop(columns=['ID', 'Date House was Sold','No of Times Visited'], axis=1, inplace=True)
```

```python
house_pricing.columns
```

```python
#Filling missing values in no of bathrooms with mean imputation
house_pricing['No of Bathrooms'] = house_pricing['No of Bathrooms'].fillna(house_pricing['No of Bathrooms'].mean())
```

```python
#Filling missing values in Flat Area (in Sqft) with mean imputation
house_pricing['Flat Area (in Sqft)'] = house_pricing['Flat Area (in Sqft)'].fillna(house_pricing['Flat Area (in Sqft)'].mean())
```

```python
#Filling missing values in Lot Area (in Sqft) with mean imputation
house_pricing['Lot Area (in Sqft)'] = house_pricing['Lot Area (in Sqft)'].fillna(house_pricing['Lot Area (in Sqft)'].mean())
```

```python
#Filling missing values in Area of the House from Basement(in Sqft) with mean imputation
house_pricing['Area of the House from Basement (in Sqft)'] = house_pricing['Area of the House from Basement (in Sqft)'].fillna(house_pricing['Area of the House from Basement (in Sqft)'].mean())
```

```python
#Filling missing values in Zipcode with median imputation
house_pricing['Zipcode'] = house_pricing['Zipcode'].fillna(house_pricing['Zipcode'].median())
```

```python
#Filling missing values in Latitude with median imputation
house_pricing['Latitude'] = house_pricing['Latitude'].fillna(house_pricing['Latitude'].median())
```

```python
#Filling missing values in Longitude with median imputation
house_pricing['Longitude'] = house_pricing['Longitude'].fillna(house_pricing['Longitude'].median())
```

```python
#Filling missing values in Living Area after Renovation (in Sqft) with mean imputation
house_pricing['Living Area after Renovation (in Sqft)'] = house_pricing['Living Area after Renovation (in Sqft)'].fillna(house_pricing['Living Area after Renovation (in Sqft)'].mean())
```

```python
#Dropping missing values in sale price
house_pricing = house_pricing.dropna(subset=['Sale Price'])
```

####Outlier Removal

```python
Q1 = house_pricing['No of Bedrooms'].quantile(0.25)
Q3 = house_pricing['No of Bedrooms'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(Q1, Q3, IQR, lower_bound, upper_bound)
```

```python
house_pricing['No of Bedrooms'] = house_pricing['No of Bedrooms'].clip(lower_bound, upper_bound)
sns.boxplot(data = house_pricing['No of Bedrooms'])
```

```python
num_cols = ['No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)', 'No of Floors', 'Overall Grade',
    'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)', 'Age of House (in Years)', 'Zipcode',
    'Latitude', 'Longitude', 'Living Area after Renovation (in Sqft)', 'Lot Area after Renovation (in Sqft)']

for col in num_cols:
    Q1 = house_pricing[col].quantile(0.25)
    Q3 = house_pricing[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    house_pricing[col] = house_pricing[col].clip(lower_bound, upper_bound)
```

```python
Q1 = house_pricing['Sale Price'].quantile(0.25)
Q3 = house_pricing['Sale Price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(Q1, Q3, IQR, lower_bound, upper_bound)
```

```python
house_pricing['Sale Price'] = house_pricing['Sale Price'].clip(lower_bound, upper_bound)
sns.boxplot(data = house_pricing['Sale Price'])
```

####Scaling Numerical Variables

```python
#Scaling using techniques like Min-Max scaling or Standard scaling
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler(feature_range = (0,1))
num_cols = ['No of Bedrooms', 'No of Bathrooms','Flat Area (in Sqft)', 'Lot Area (in Sqft)', 'No of Floors', 'Overall Grade',
            'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)', 'Age of House (in Years)', 'Renovated Year',
            'Zipcode', 'Latitude', 'Longitude', 'Living Area after Renovation (in Sqft)', 'Lot Area after Renovation (in Sqft)']
house_pricing[num_cols]=min_max_scaler.fit_transform(house_pricing[num_cols])
house_pricing
```

####Train-Test Split

```python
from sklearn.model_selection import train_test_split

x = house_pricing.drop(columns = ['Sale Price'])
y = house_pricing['Sale Price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

print("Shape of X_train:", x_train.shape)
print("Shape of X_test:", x_test.shape)
print("Shape of Y_train:", y_train.shape)
print("Shape of Y_test:", y_test.shape)
```
